```python
#pip install numpy pandas scikit-learn tensorflow
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

def dna_one_hot_encode(seq):
    mapping = {"A": [1,0,0,0], "C": [0,1,0,0], "G": [0,0,1,0], "T": [0,0,0,1]}
    seq = seq.upper().replace(" ", "").replace("\t", "")
    try:
        return np.array([mapping[char] for char in seq])
    except KeyError as e:
        print(f"Karakter tidak valid '{e.args[0]}' ditemukan di sekuens: {seq}")
        raise
        
# Contoh data DNA (panjang sekuens = 60)
sequences = [
    "ATCGTGACGTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGC",
    "GCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAG",
    "CGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTAC",
    "TGCATGCATGCATGCATGCATGCATGCATGCATGCATGCATGCATGCATGCATGCAT",
]
labels = ["positive", "positive", "negative", "negative"]


df["sequence"] = df["sequence"].str.upper().str.replace(r"[^ACGT]", "", regex=True)
sequences = df["sequence"]
labels = df["label"]
encoded_seqs = np.array([dna_one_hot_encode(seq) for seq in sequences])
print("Encoded shape:", encoded_seqs.shape) # (jumlah_sampel, panjang, 4)

  
le = LabelEncoder()
y = le.fit_transform(labels)
y_cat = to_categorical(y)

print(encoded_seqs)
```

```python
model = Sequential([
    Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(encoded_seqs.shape[1], 4)),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(100, activation='relu'),
    Dropout(0.5),
    Dense(y_cat.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(encoded_seqs, y_cat, test_size=0.25, random_state=42)

#model.fit(X_train, y_train, epochs=20, batch_size=2, validation_data=(X_test, y_test))
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns

# TRAIN MODEL & simpan history
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=2,
    validation_data=(X_test, y_test)
)

# EVALUASI
loss, acc = model.evaluate(X_test, y_test)
print(f"\n✅ Akurasi Test: {acc * 100:.2f}%")

# PLOTTING LOSS & AKURASI
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# CONFUSION MATRIX
from sklearn.metrics import classification_report
import numpy as np

# Prediksi kelas
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred_classes)
labels = le.classes_  # gunakan label asli jika ada
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()

# Laporan Klasifikasi
print("\n📊 Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=labels))
```

```python
import numpy as np
import networkx as nx

# Nilai bobot antara pasangan basa DNA (arbitrer atau bisa berdasarkan statistik biologis)
nuc_weights = {
    ('A', 'C'): 1, ('A', 'G'): 2, ('A', 'T'): 3,
    ('C', 'G'): 1, ('C', 'T'): 2,
    ('G', 'T'): 1,
}

# Buat simetri
for (a, b), w in list(nuc_weights.items()):
    nuc_weights[(b, a)] = w

def dna_to_graph(seq):
    G = nx.Graph()
    for i, base in enumerate(seq):
        G.add_node(i, label=base)
        if i > 0:
            prev = seq[i-1]
            curr = base
            weight = nuc_weights.get((prev, curr), 5)  # default weight jika tidak ada
            G.add_edge(i-1, i, weight=weight)
    return G

def mst_cost(seq):
    G = dna_to_graph(seq)
    mst = nx.minimum_spanning_tree(G, weight='weight')
    return mst.size(weight='weight')  # total bobot MST

# Contoh sekuens
sequencesTes = [
    "ACGTACGT",
    "TTTTGGGG",
    "AGCTAGCT"
]

features = [mst_cost(seq) for seq in sequencesTes]
print("Fitur MST:", features)
```

```python
import numpy as np

# Definisi nukleotida dan klasifikasi transisi/transversi
purines = {'A', 'G'}
pyrimidines = {'C', 'T'}

def kimura_distance(nuc1, nuc2):
    if nuc1 == nuc2:
        return 0.0
    elif (nuc1 in purines and nuc2 in purines) or (nuc1 in pyrimidines and nuc2 in pyrimidines):
        return 1.0  # transition
    else:
        return 2.0  # transversion
```

```python
def sequence_to_kimura_matrix(seq):
    seq = seq.upper()
    n = len(seq)
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            matrix[i][j] = kimura_distance(seq[i], seq[j])
    return matrix
```

```python
#GCN
import torch
from torch_geometric.data import Data
import numpy as np

# Mapping nukleotida ke integer
nuc2idx = {'A': 0, 'C': 1, 'G': 2, 'T': 3}

def kimura_distance(n1, n2):
    if n1 == n2:
        return 0.0
    purines = {'A', 'G'}
    pyrimidines = {'C', 'T'}
    if (n1 in purines and n2 in purines) or (n1 in pyrimidines and n2 in pyrimidines):
        return 1.0
    return 2.0

def sequence_to_graph(seq, window=2):
    seq = seq.upper()
    n = len(seq)
    
    # Node features: simple one-hot
    x = torch.eye(4)[[nuc2idx[base] for base in seq]]  # shape [n, 4]
    
    # Edge index and weights
    edge_index = []
    edge_weight = []
    
    for i in range(n):
        for offset in range(-window, window+1):
            j = i + offset
            if 0 <= j < n and i != j:
                edge_index.append([i, j])
                d = abs(i - j)
                weight = kimura_distance(seq[i], seq[j]) * (1 / d)
                edge_weight.append(weight)
    
    edge_index = torch.tensor(edge_index).t().contiguous()  # shape [2, num_edges]
    edge_weight = torch.tensor(edge_weight, dtype=torch.float)
    
    return Data(x=x, edge_index=edge_index, edge_attr=edge_weight)


from sklearn.preprocessing import LabelEncoder
from torch_geometric.data import DataLoader

# asumsi df["sequence"], df["label"] sudah ada
le = LabelEncoder()
y = le.fit_transform(df["label"])

graphs = []
for i, row in df.iterrows():
    g = sequence_to_graph(row["sequence"])
    g.y = torch.tensor([y[i]], dtype=torch.long)
    graphs.append(g)

# Bagi train/test
from sklearn.model_selection import train_test_split
train_graphs, test_graphs = train_test_split(graphs, test_size=0.2, random_state=42)

train_loader = DataLoader(train_graphs, batch_size=16)
test_loader = DataLoader(test_graphs, batch_size=16)
```

```python
import torch.nn.functional as F
from torch_geometric.nn import GraphConv, global_mean_pool
from torch.nn import Linear
import torch.nn as nn

class GCNClassifier(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super().__init__()
        self.conv1 = GraphConv(input_dim, hidden_dim)
        self.conv2 = GraphConv(hidden_dim, hidden_dim)
        self.lin = Linear(hidden_dim, num_classes)
    
    def forward(self, data):
        x, edge_index, edge_attr, batch = data.x, data.edge_index, data.edge_attr, data.batch
        x = self.conv1(x, edge_index, edge_attr)
        x = F.relu(x)
        x = self.conv2(x, edge_index, edge_attr)
        x = F.relu(x)
        x = global_mean_pool(x, batch)  # pooling untuk graph-level output
        x = self.lin(x)
        return F.log_softmax(x, dim=1)
```
